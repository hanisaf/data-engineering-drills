# Drill: Keyword Filter Stream

Connect to a socket text stream and filter lines that contain any of the watch keywords `["@hanisaf", "#uga", "#python"]`.  
Save the matching lines to text files and print them to the console.

## Setup

Before running the streaming cell, open a terminal and start a socket server that forwards your keyboard input on port 7777:

```bash
# Option 1 -- netcat (simplest)
nc -lk 7777

# Option 2 -- Python equivalent of SocketTool.java
python3 socket_tool.py 7777
```

Once the socket server is running, execute the streaming cell below.  
Type lines such as:
```
Hello world
I love #uga football
@hanisaf teaches #java
This line should not match
```
The streaming job will print only the matching lines every 10 seconds.  
**Interrupt the kernel** (■ button or Kernel → Interrupt) to stop the stream.

# Your Task

The stub below has a filter function `f` that always returns `False` (no lines match) -- exactly like the Java start code.  

**Implement `f`** so it returns `True` when `line` contains at least one of the `watch_keywords`.

````python
# ── Stub (your task: implement the filter function f) ────────────────────
ssc = StreamingContext(sc, 10)   # 10-second batch interval

lines = ssc.socketTextStream(STREAM_HOST, 7777)

watch_keywords = ["@hanisaf", "#uga", "#python"]

# TODO: implement f so it returns True when line contains any watch keyword
f = lambda line: False

filtered = lines.filter(f)
filtered.saveAsTextFiles("out/filter", "txt")
filtered.pprint()

print("Streaming started -- interrupt the kernel to stop.")
ssc.start()
ssc.awaitTermination()
ssc.stop(stopSparkContext=False)
````
