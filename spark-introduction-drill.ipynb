# Drill -- The Birthday Paradox

## Background

In a group of 23 randomly chosen people, the probability that at least two share a birthday exceeds **50%**. This seems surprising -- hence the name "paradox".

## Task

Use Spark to **empirically verify** this via simulation:

1. Generate `num_trials = 1000` trials. Each trial is a list of `group_size = 23` random integers in [1, 365] representing birthdays.
2. For each trial, detect whether a **collision** exists -- i.e., two people share a birthday -- by checking if `len(set(group)) < group_size`.
3. Count trials that had a collision.
4. Report: `collision_probability = collisions / num_trials`

Use `sc.parallelize()` to distribute the trials across **4 partitions**.

> **Expected result**: ~50–55% for group_size=23.

Reference: https://en.wikipedia.org/wiki/Birthday_problem

```python
import random

DAYS = 365
num_trials = 1000
group_size = 23

# Generate num_trials groups of group_size random birthdays
data = [[random.randint(1, DAYS) for _ in range(group_size)] for _ in range(num_trials)]
print("Preview:", data[:2])
```

```python
rdd = sc.parallelize(data, 4)

def has_collision(group):
    return 0 if len(set(group)) == len(group) else 1  # 1 = collision found

# collisions = rdd.map(....).reduce(....)
# probability = collisions / num_trials
```
